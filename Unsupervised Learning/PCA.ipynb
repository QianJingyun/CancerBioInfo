In this file, we performed PCA on the 58 features selected from K_mean clustering. See 'KMeans.ipynb' for how to obtain desired table from raw data. We presented:
1. PCA plots on 58 features.

```python
# load data file
data = pd.read_csv('best_table.csv')
data
```

```python
X = data.iloc[:,1:]
y = data['PATH_N_STAGE']
```

### PCA with 58-genes set

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X)
```

```python
plt.scatter(principal_components[:,0], principal_components[:,1], c= y, s = 20, alpha = 0.7)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA plot using 58 variables')
plt.show()
```
